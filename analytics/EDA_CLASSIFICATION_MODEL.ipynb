## 1.1 Dataset Info

O dataset apresenta 60000 observações não nulas. O tratamento de valores ausentes e infinitos foi feito tratando-os como zero. Assim, se uma observação apresenta uma variação percentual infinita (pois não há observação anterior), então considerou-se a variação da feature como zero. É uma aproximação que pode ser mudada a critério do desenvolvimento do modelo.

## 1.2 Tratamento de Outliers

Retirar tudo que for maior do que 3 desvios padrão.

```python
'''
trava muito
for column in dataset.columns.tolist()[:-2]:
    fig = px.box(dataset, y=column)
    fig.show()
''';
```

```python
m=3
for column in dataset.columns.tolist()[:-2]:
    dataset = dataset[np.abs(dataset[column]-dataset[column].mean()) <= (m*dataset[column].std())]
```

# 5. Modelo de Classificação

```python
dataset.sort_values("fiscal_date", inplace=True)
```

```python
dataset['label'] = dataset['return'] > 0.05
dataset['label'] = dataset['label'].astype(int)
```

```python
X_train = dataset[dataset['fiscal_date'] < date(2017,12,31)].copy()
X_test = dataset[dataset['fiscal_date'] >= date(2017,12,31)].copy()
```

```python
y_train = X_train['label']
y_test = X_test['label']
```

```python
X_train.drop(["label"], axis=1, inplace=True)
X_test.drop(["label"], axis=1, inplace=True)
```

```python
y_train[y_train == 0].hist();
y_train[y_train == 1].hist();

print("% 0 treino:",len(y_train[y_train == 0])/len(y_train))
print("% 1 treino:", len(y_train[y_train == 1])/len(y_train))
```

```python
y_test[y_test == 0].hist();
y_test[y_test == 1].hist();

print("% 0 teste:",len(y_test[y_test == 0])/len(y_test))
print("% 1 teste:", len(y_test[y_test == 1])/len(y_test))
```

```python
test_fiscal_date = X_test['fiscal_date']
test_ticker = X_test['Ticker']
test_return = X_test['return']
```

```python
X_train.drop(['fiscal_date','Ticker', 'return'], axis=1, inplace=True)
```

```python
X_test.drop(['fiscal_date','Ticker', 'return'], axis=1, inplace=True)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_estimators=200, learning_rate=0.05)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
target_names = ['class 0', 'class 1']
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
X_test['Ticker'] = test_ticker
X_test['fiscal_date'] = test_fiscal_date
X_test['y_pred'] = y_pred
X_test['return'] = test_return
```
